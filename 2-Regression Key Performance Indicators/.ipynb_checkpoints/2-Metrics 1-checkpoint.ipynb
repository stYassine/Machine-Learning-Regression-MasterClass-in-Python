# //////////////////////////////////////////
# /// 1 - Mean Absolute Error (MEA)
# //////////////////////////////////////////

- MAE is Obtained by calculating the Absolute difference between the model Predictions and the true (actual) values.
- why it's called "Mean" ?
- why it's called "Absolute" ? : because we're going to calculate the Absolute value.
<br><br>
    <b>Equation Explanation</b>
<img src="./assets/2-Mean Absolute Error.png" />

- i'm going to take every point with subset i, calculate the difference between yi(true values) minus Å·i(predictions),
i'm going to calculate the Absolute Value,
then i'm gonna obtain the Average, average of all these points, which is straight forward
i'm gonna sum up all these errors and divide by the overall number of samples 
<br><br>
- MAE is a measure of the average magnitude of error generated by regression model
- if MAE is zero, this indicated that the model predictions are perfect.

# //////////////////////////////////////////
# /// 2- Mean Square Error (MSE)
# //////////////////////////////////////////

(3-Mean Square Error)
- it's pretty much the same as MAE, but instead of Obtaining the Absolute value, i'm going to take the Error and Square it
- that's why it's called Mean SQuared Error
<br><br>
- MSE values are generally "larger" compared to the MAE why ?
because we simply Square it up, (we are taking the Residuals and Square them up)
- in case of data outliers, MSE will become much larger compared to MAE,
think of it as if we're penalizing the mistake, if we made a mistake this Error will be Squared.
<br><br>
- in MSE, error increases in a Quadratic fashion (because we're squaring it up), 
while in case of MAE it increases in a "proportional fashion" (because we're not squaring it)
<br><br>
- in MSE, since the Error is being "Squared", any prediction Error is Heavily Penalized
<br><br>
    <b>Equation Explanation<b/>
     <img src="./assets/3-Mean Square Error.png" />
MSE is calculated by following
- 1- Calculate the residual for every data point
- 2- Calculate the squared value of the residuals (take the residuals and square them up)
- 3- Calculate the average of results from step #2

# //////////////////////////////////////////
# /// 3- Root Mean Square Error (RMSE)
# //////////////////////////////////////////

- it's the pretty simple, take the Exact same values from (MSE) and Obtain the Square Root for it
<br><br>
- The Question is Why ? why do we need it ?
- the Problem with (MSE), is that the unites are very different to the Actual data points, 
Because we're already Squaring them up, so it's hard to Compare.
- that's why if we Obtain the Square Root of this data,
then we're gonna go back to match the units of the Original data, so we can compare it
<br><br>
- Root Mean Square Error (RMSE), represents the "Standard deviation of the residuals",
in Statistics "Standard deviation" is a Measure of dispersion from the "MEAN",
which is how much i'm a Spread AWAY from the Mean.
<br><br>
- RMSE can be easily "Interpreted" compared to MSE because RMSE units match the units of the output (Original data).
<br><br>
- from Equation, i'm simply squaring it up and then Obtaining the Square root,
it's like Canceling the squaring effect.
<br><br>
- RMSE provides an estimate of how large the residuals are being dispersed,
Again it's a Standard deviation
<br><br>
    <b>Equation Explanation</b>
    <img src="./assets/4-Root Mean Square Error.png" />

- 1- Calculate the Residual for every data point
- 2- Calculate the squared Value of the residuals
- 3- Calculate the average of the squared residuals
- 4- Obtain the square root of the result

# //////////////////////////////////////////
# /// 4- Mean Absolute Percentage Error (MAPE)
# //////////////////////////////////////////

- 1 Problem with MAE Values or metrics is it can range from 0 to Infinity,
which makes it difficult to interpret the results, 
as compared to the training data. 
<br><br>
- that's why the Answer in a form of Percentge is very beneficial, because everyone can Understand it (my Error is 5%, my Error is 10%)
<br><br>
- (MAPE) is the Equivalent to (MAE) but provides the Error in a percentage form, and therefore overcomes the (MAE) limitations.
<br><br>
- (MAPE) might exhibit some limitations, why ? 
if the data point value is zero (there is a division Operation involved)
<br><br>
    <b>Equation Explanation</b>
    <img src="./assets/5-Mean Absolute Percentage Error.png" />
- take every data point
- subtract the Estimate from the True value (residual)
- divide by the Original value of the data point yi
- and then sum them up
- then get the Average divide by n
- then multiply by 100%
- that will give me a Percentage value as compared to the Actual value of the data point
- the problem here is, if point yi is 0 then i'm gonna Exhibit some problems (because i'm gonna be dividing by 0)

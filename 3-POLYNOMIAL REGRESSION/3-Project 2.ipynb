# Problem Statement
- Let's assume that you work as a consultant to a start-up company that was looking to develop a model to estimate the cost of good sold as they vary the production volume (number of units producted).
- The startup gathered data and has asked you to develop a model to predict its cost vs. number of units sold.
- You thought a polynomial regression model might be a good candidate.

# Step 1 : Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step 2 : Import Dataset

```python
dataset = pd.read_csv('./EconomiesOfScale.csv')
```

```python
dataset.head()
```

# Step 3 : Visualize Dataset

```python
sns.jointplot(x='Number of Units', y='Manufacturing Cost', data=dataset)
```

```python
sns.lmplot(x='Number of Units', y='Manufacturing Cost', data=dataset)
```

```python
# Reverse Parameters
sns.jointplot(y='Number of Units', x='Manufacturing Cost', data=dataset)
```

```python
sns.pairplot(dataset)
```

# Step 4 : Create Training Dataset

```python
X = dataset[['Number of Units']]
y = dataset[['Manufacturing Cost']]
```

```python
X_train = X
y_train = y
```

X_train.shape

```python
y_train.shape
```

# Step 5 : Train The Model

### 5.1 - Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
linearRegression = LinearRegression(fit_intercept=False)
```

```python
linearRegression.fit(X_train, y_train)
```

```python
print('Model Coeff(m) : ', linearRegression.coef_)
print('Model Coeff(b) : ', linearRegression.intercept_)
```

### Step 5.2 : Visualize the Result

```python
predict_X_train = linearRegression.predict(X_train)
plt.scatter(X_train, y_train, color='grey')
plt.plot(X_train, predict_X_train, color='red')
plt.xlabel('Number of Units [in Millions]')
plt.ylabel('Cost Per unit sold ')
plt.title('Units Cost VS Number of Units [Linear]')
```

```python
### fit_intercept = True
# Model Coeff(m) :  [[-6.0333683]]
# Model Coeff(b) :  [67.03904193]
###  fit_intercept = False
# Model Coeff(m) :  [[7.72782912]]
# Model Coeff(b) :  0.0
```

### 4.2 - Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
polynomialFeatures = PolynomialFeatures(degree=5)
```

```python
X_columns = polynomialFeatures.fit_transform(X_train)
```

```python
X_columns
```

```python
regressor = LinearRegression()
regressor.fit(X_columns, y_train)
```

```python
print('Model Coeff(m) : ', regressor.coef_)
print('Model Coeff(b) : ', regressor.intercept_)
```

### Step 5.3 : Visualize the Result

```python
y_predict = regressor.predict(X_columns)
```

```python
plt.scatter(X_train, y_train, color='grey')
plt.plot(X_train, y_predict, color='red')
plt.xlabel('Number of Units [in Millions]')
plt.ylabel('Cost Per unit sold ')
plt.title('Units Cost VS Number of Units [Poly Order = 2]')
```

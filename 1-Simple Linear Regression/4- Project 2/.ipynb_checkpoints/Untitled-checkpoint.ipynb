# Problem Statement
- You have been hired as a consultant to a major Automotive Manufacturer and you have been tasked to develop a model to predict the impact of increasing the vehicle horsepower (HP)
on fuel economy (Mileage Per Gallon (MPG)).
- You gatherd the Data:
	- Independant variable X: Vehicle Horse Power
	- Dependant variable Y: Mileage Per Gallon (MPG)

# 1 - Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# 2 - Import Data

```python
data = pd.read_csv('./FuelEconomy.csv')
```

```python
data.head()
```

# 3 - Visualization

```python
sns.jointplot(x = 'Horse Power', y='Fuel Economy (MPG)', data = data)
```

```python
sns.jointplot(x = 'Fuel Economy (MPG)', y='Horse Power', data = data)
```

```python
sns.pairplot(data)
```

```python
sns.lmplot(x = 'Horse Power', y='Fuel Economy (MPG)', data = data)
```

```python
sns.lmplot(x = 'Fuel Economy (MPG)', y='Horse Power', data = data)
```

# 4 - Split Data

```python
X = data[['Horse Power']]
```

```python
y = data[['Fuel Economy (MPG)']]
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

# 5 - Train the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor = LinearRegression(fit_intercept=True)
regressor.fit(X_train, y_train)
```

```python
print('Linear Model Coeff (m) : ', regressor.coef_)
print('Linear Model Coeff (b) : ', regressor.intercept_)
```

# 6 - Test the Model

```python
y_predict = regressor.predict(X_test)
```

```python
y_predict
```

```python
X_train_predict = regressor.predict(X_train)
plt.scatter(X_train, y_train, color = 'grey')
plt.plot(X_train, X_train_predict, color = 'blue')
plt.xlabel('Horse Power (HP)')
plt.ylabel('MPG')
plt.title('HP vs. MPG (Training Set)')
```
